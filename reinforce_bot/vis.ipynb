```python
import bz2
import pickle
import _pickle as cPickle
def decompress_pickle(file):
    data = bz2.BZ2File(file, 'rb')
    data = cPickle.load(data)
    return data
```

```python
eps_history = decompress_pickle("model/eps_history.pbz2")
scores = decompress_pickle("model/scores_history.pbz2")
```
